```python
from sklearn.preprocessing import OrdinalEncoder

# Preprocesamiento
print("\nPreprocesando los datos...")

# Verificar las columnas existentes
print("\nColumnas en el conjunto de entrenamiento:")
print(train_data.columns)

# Eliminar 'class' del conjunto de prueba
if 'class' in test_data.columns:
    test_data = test_data.drop(columns=['class'])
    print("Eliminada columna 'class' del conjunto de prueba.")

# Eliminar columnas innecesarias si existen
if 'Unnamed: 0' in train_data.columns:
    train_data = train_data.drop(columns=['Unnamed: 0'])
if 'Unnamed: 0' in test_data.columns:
    test_data = test_data.drop(columns=['Unnamed: 0'])
```

```python
# Separar características y etiquetas
X = train_data.drop(columns=['class'])
y = train_data['class']

# Codificar las etiquetas
label_encoder = LabelEncoder()
y_encoded = label_encoder.fit_transform(y)  # E -> 0, P -> 1
print(f"\nEtiquetas transformadas: {list(label_encoder.classes_)} -> {list(range(len(label_encoder.classes_)))}")

# Codificar las características con OrdinalEncoder
ordinal_encoder = OrdinalEncoder()

X_encoded = pd.DataFrame(
    ordinal_encoder.fit_transform(X),
    columns=X.columns
)

test_data_encoded = pd.DataFrame(
    ordinal_encoder.transform(test_data),
    columns=test_data.columns
)

print("\nColumnas utilizadas para entrenar:")
print(X_encoded.columns.tolist())
print("\nPrimeras filas del conjunto de entrenamiento codificado:")
print(X_encoded.head())
```

```python
# Asegurar que las columnas coincidan
test_data_encoded = test_data_encoded.reindex(columns=X_encoded.columns, fill_value=0)
print(f"\nColumnas después de codificación de variables categóricas: {X_encoded.columns.tolist()}")

X_train, X_val, y_train, y_val = train_test_split(X_encoded, y_encoded, test_size=0.2, random_state=42, stratify=y_encoded)
print(f"Tamaño del conjunto de entrenamiento: {X_train.shape}")
print(f"Tamaño del conjunto de validación: {X_val.shape}")
```

# Selection and model training

```python
# Modelos y evaluación
metrics = []

# Modelo 1: k-NN
print("\nEntrenando modelo k-NN...")
knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(X_train, y_train)
y_pred_knn = knn.predict(X_val)
accuracy_knn = accuracy_score(y_val, y_pred_knn)
f1_knn = f1_score(y_val, y_pred_knn, average='weighted')
metrics.append(('k-NN', accuracy_knn, f1_knn))
print(f"k-NN - Accuracy: {accuracy_knn}, F1-Score: {f1_knn}")

# Modelo 2: Árbol de Decisión
print("\nEntrenando modelo Árbol de Decisión...")
decision_tree = DecisionTreeClassifier(random_state=42)
decision_tree.fit(X_train, y_train)
y_pred_tree = decision_tree.predict(X_val)
accuracy_tree = accuracy_score(y_val, y_pred_tree)
f1_tree = f1_score(y_val, y_pred_tree, average='weighted')
metrics.append(('Decision Tree', accuracy_tree, f1_tree))
print(f"Árbol de Decisión - Accuracy: {accuracy_tree}, F1-Score: {f1_tree}")

# Modelo 3: SVM
print("\nEntrenando modelo SVM...")
svm = SVC(kernel='linear', random_state=42)
svm.fit(X_train, y_train)
y_pred_svm = svm.predict(X_val)
accuracy_svm = accuracy_score(y_val, y_pred_svm)
f1_svm = f1_score(y_val, y_pred_svm, average='weighted')
metrics.append(('SVM', accuracy_svm, f1_svm))
print(f"SVM - Accuracy: {accuracy_svm}, F1-Score: {f1_svm}")
```

# Prediction and evaluation

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OrdinalEncoder, LabelEncoder
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC
import pandas as pd
import numpy as np

# Supongamos que `train_data` y `test_data` ya están cargados

# Preprocesamiento
X = train_data.drop(columns=['class'])
y = train_data['class']

label_encoder = LabelEncoder()
y_encoded = label_encoder.fit_transform(y)

ordinal_encoder = OrdinalEncoder()
X_encoded = pd.DataFrame(
    ordinal_encoder.fit_transform(X),
    columns=X.columns
)

test_data_encoded = pd.DataFrame(
    ordinal_encoder.transform(test_data.drop(columns=['class'] if 'class' in test_data.columns else [])),
    columns=test_data.columns
)

# Dividir en conjunto de entrenamiento y validación
X_train, X_val, y_train, y_val = train_test_split(X_encoded, y_encoded, test_size=0.2, random_state=42, stratify=y_encoded)

# Entrenar los modelos
knn = KNeighborsClassifier(n_neighbors=5)
decision_tree = DecisionTreeClassifier(random_state=42)
svm = SVC(kernel='linear', random_state=42)

knn.fit(X_train, y_train)
decision_tree.fit(X_train, y_train)
svm.fit(X_train, y_train)

# Generar predicciones con todos los modelos y mostrar resultados
models = [
    (knn, "k-NN"),
    (decision_tree, "Decision Tree"),
    (svm, "SVM")
]

print("\nGenerando predicciones finales para todos los modelos...")
for model, name in models:
    print(f"\nModelo: {name}")
    test_predictions = model.predict(test_data_encoded)  # Predicciones
    test_predictions_labels = label_encoder.inverse_transform(test_predictions)  # Convertir a etiquetas originales

    # Mostrar un resumen de las predicciones
    unique, counts = np.unique(test_predictions_labels, return_counts=True)
    results_summary = dict(zip(unique, counts))
    print(f"Resumen de predicciones: {results_summary}")

    # Guardar el archivo CSV para cada modelo
    submission = pd.DataFrame({
        'ID': test_data.index + 1,  # Ajustar ID para que empiece desde 1
        'Edible': test_predictions_labels
    })

    submission_file_path = f'submission_{name.lower().replace(" ", "_")}.csv'
    submission.to_csv(submission_file_path, index=False)
    print(f"Archivo guardado: {submission_file_path}")
```

```python
# Hacer predicciones en el conjunto de validación
y_val_pred = decision_tree.predict(X_val)

# Crear la matriz de confusión
conf_matrix = confusion_matrix(y_val, y_val_pred, labels=[0, 1])

# Visualizar la matriz de confusión
disp = ConfusionMatrixDisplay(confusion_matrix=conf_matrix, display_labels=label_encoder.classes_)
disp.plot(cmap="Blues")
plt.title("Matriz de Confusión (Árbol de Decisión)")
plt.show()
```

- Verdaderos Positivos (TP): Casos correctamente clasificados como positivos.
- Verdaderos Negativos (TN): Casos correctamente clasificados como negativos.
- Falsos Positivos (FP): Casos negativos clasificados incorrectamente como positivos.
- Falsos Negativos (FN): Casos positivos clasificados incorrectamente como negativos.
