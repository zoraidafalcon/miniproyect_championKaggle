# Importar librerias

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OrdinalEncoder, LabelEncoder
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, f1_score, confusion_matrix, ConfusionMatrixDisplay
```

# 1. Cargar los datos

```python
print("\nCargando los datos...")
train_data = pd.read_csv('train.csv')
test_data = pd.read_csv('test.csv')
print("Datos cargados correctamente.")
print(f"\nEstructura del conjunto de entrenamiento:\n{train_data.info()}")
print(f"\nEstructura del conjunto de prueba:\n{test_data.info()}")
```

# 2. Preprocesamiento

```python
print("\nPreprocesando los datos...")

# Eliminar columnas innecesarias si existen
if 'Unnamed: 0' in train_data.columns:
    train_data = train_data.drop(columns=['Unnamed: 0'])
if 'Unnamed: 0' in test_data.columns:
    test_data = test_data.drop(columns=['Unnamed: 0'])
```

```python
# Separar características y etiquetas
X = train_data.drop(columns=['class'])
y = train_data['class']

# Codificar etiquetas
label_encoder = LabelEncoder()
y_encoded = label_encoder.fit_transform(y)  # E -> 0, P -> 1
print(f"Etiquetas transformadas: {list(label_encoder.classes_)} -> {list(range(len(label_encoder.classes_)))}")

# Codificar características categóricas
ordinal_encoder = OrdinalEncoder()
X_encoded = pd.DataFrame(ordinal_encoder.fit_transform(X), columns=X.columns)
test_data_encoded = pd.DataFrame(
    ordinal_encoder.transform(test_data.drop(columns=['class'] if 'class' in test_data.columns else [])),
    columns=test_data.columns
)
```

```python
# Dividir en conjunto de entrenamiento y validación
X_train, X_val, y_train, y_val = train_test_split(X_encoded, y_encoded, test_size=0.2, random_state=42, stratify=y_encoded)

print(f"Tamaño del conjunto de entrenamiento: {X_train.shape}")
print(f"Tamaño del conjunto de validación: {X_val.shape}")
```

# 3. Entrenamiento de modelos

```python

print("\nEntrenando modelos...")
```

```python
# Definir modelo de KNN
knn = KNeighborsClassifier(n_neighbors=5)
# Entrenar modelo
knn.fit(X_train, y_train)
```

```python
# Definir modelo de DT
decision_tree = DecisionTreeClassifier(random_state=42)
# Entrenar modelo
decision_tree.fit(X_train, y_train)
```

```python
# Definir modelo de SVM 
svm = SVC(kernel='linear', random_state=42)
# Entrenar modelo
svm.fit(X_train, y_train)
```

# 4. Evaluar modelos en validación

```python
models = [
    (knn, "k-NN"),
    (decision_tree, "Decision Tree"),
    (svm, "SVM")
]

print("\nEvaluando modelos en conjunto de validación...")
for model, name in models:
    y_val_pred = model.predict(X_val)
    acc = accuracy_score(y_val, y_val_pred)
    f1 = f1_score(y_val, y_val_pred, average='weighted')
    print(f"\nModelo: {name}")
    print(f"Accuracy: {acc:.4f}, F1-Score: {f1:.4f}")

    # Matriz de confusión
    conf_matrix = confusion_matrix(y_val, y_val_pred, labels=[0, 1])
    disp = ConfusionMatrixDisplay(confusion_matrix=conf_matrix, display_labels=label_encoder.classes_)
    disp.plot(cmap="Blues")
    plt.title(f"Matriz de Confusión: {name}")
    plt.show()
```

# 5. Generar predicciones finales en el conjunto de prueba

```python
# Generar predicciones finales con el modelo seleccionado
selected_model, selected_model_name = decision_tree, "Decision Tree"

# Realizar predicciones
test_predictions = selected_model.predict(test_data_encoded)
test_predictions_labels = label_encoder.inverse_transform(test_predictions)

# Crear el archivo de predicciones en el formato correcto
submission = pd.DataFrame({
    'ID': range(1, len(test_predictions_labels) + 1),  # IDs consecutivos comenzando en 1
    'Edible': test_predictions_labels  # Predicciones ('E' o 'P')
})

# Guardar archivo CSV
submission_file_path = 'submission_corrected.csv'
submission.to_csv(submission_file_path, index=False)
print(f"Archivo guardado para Kaggle: {submission_file_path}")
```
